1. Write a program to create a dictionary of Bengali words with values as their English
translation. Provide user with an option to look it up!

```python
dictionary = {
    "Banana" : "kola",
    "Orange" : "lebu",
    "Name" : "nam",
    "chair" : "aram kedara"
}

name = input("enter a bengali word which is in the list: ")

print(dictionary[name])
```

2. Write a program to input eight numbers from the user and display all the unique
numbers (once).

```python
user_input = input("enter the number space separated: ")

numbers = user_input.split()       # at first split it out

inputs = set(numbers)              # then convert it into set to remove duplicates

print(inputs)
```

4. What will be the length of following set s:

s = set()

s.add(20)

s.add(20.0)

s.add('20') # length of s after these operations?

```python
s = set()
s.add(20)
s.add(20.0)
s.add('20') # length of s after these operations?

len(s)   # 2, because 20 and 20.0 are considered the same in Python sets, while '20' is a different string.
```

6. Create an empty dictionary. Allow 4 friends to enter their favorite language as
value and use key as their names. Assume that the names are unique.

```python
dc = {}

n = int(input("enter no of items: "))

for i in range(n):
    key = input("enter key : ")
    value = input("enter value : ")
    dc[key] = value

print(dc)
```
