Dictionary is collection of key-value pair

It is unordered

It is mutable

It is Indexed

Cannot contain duplicate keys

```python
# here arr is a dictionary

arr = {
    "key" : "value",
    "fruit" : "banana",
    "marks" : 82,
    "list" : [1, 2, 3, 4, 5] 
}
```

```python
arr["key"] # this will give us the value of key
```

```python
arr["list"] # this will give us the value of list
```

```python
arr
```

```python
arr = {
    "key" : "value",
    "fruit" : "banana",
    "marks" : 82,
    "key" : "value1", # this will silently overwrite the previous value of key
    "list" : [1, 2, 3, 4, 5] 
}
```

```python
arr
```

Dictionary Methods

```python
arr.items() # this will give us a list of tuples where each tuple is a key-value pair of the dictionary
```

```python
arr.keys() # this will give us a list of keys of the dictionary
```

```python
arr.update({"fruit" : "apple"}) # this will update the value of fruit to apple
arr.items()
```

```python
arr.get("marks") # key dile value pabi simple.
```

Sets in python

set is a collection of non-repetitive elements.

```python
s = set() # this will create an empty set    no repetition allowed in set
s.add(1)
s.add(2)

# or s = {1, 2} # this will also create a set with 1 and 2 as elements
```

```python
s
```

Properties of Sets

unordered --> element's order does't matter.

unindexed --> cannot access elements by index.

immuteable --> no way to change items in sets.

no duplication --> set cannot contain duplicate values.

Operations in sets

```python
s = {1, 8, 2, 3}
```

```python
len(s) # this will give us the number of elements in the set
```

```python
s.remove(8) # this will remove 8 from the set and update it.
s
```

```python
s.pop() # this will remove and return an arbitrary element from the set
s
```

```python
s.clear() # this will remove all the elements from the set and make it empty
s
```

```python
s = {1, 8, 2, 3}
```

```python
s.union({8,11}) # return a new set with all the elements from both sets, duplicates are removed
```

```python
s.intersection({8, 11}) # return a new set with only the elements that are common in both sets
```
