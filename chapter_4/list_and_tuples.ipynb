List methods

```python
num = [1, 8, 7, 2, 21, 15]
```

```python
num.sort()  # it sorts the list in ascending order.
num
```

```python
num.reverse()  # it reverses the list.
num
```

```python
num.append(8)  # it adds an element at the end of the list.
num
```

```python
num.insert(3,51) # it adds an element at the specified index.
num              # this will add 51 at index 3
```

```python
num
```

```python
num.pop(2) # deletes the element at index 2 and returns the list
```

```python
num
```

```python
num.remove(51) # it removes the first occurrence of the specified element from the list.
num
```

Tuples in python

a tuple is a immuteable data type in python

```python
a = () # this is an empty tuple.
type(print(a))
```

```python
a = (1,) # tuple with only one element needs a comma
```

```python
a
```

```python
a = (1, 4, 7, 7) # tuple can also contain duplicate elements.
```

```python
a
```

Tuple methods

```python
a.count(7) # it counts the number of occurrences of 7 in the tuple.
```

```python
a.index(7) # it returns the index of the first occurrence of 7 in the tuple.
```
