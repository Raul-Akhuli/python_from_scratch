4. Write a program to sum a list with 4 numbers.

```python
arr = [10, 20, 30, 40, 50]

print(sum(arr)) # sum of all the elements in the list
print(max(arr)) # maximum element in the list
print(min(arr)) # minimum element in the list
print(len(arr)) # length of the list
arr.append(60) # add an element to the end of the list
print(arr)
```
